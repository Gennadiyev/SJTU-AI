# Using `matplotlib`

```python
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')

import numpy as np
import matplotlib.pyplot as plt
plt.figure(num = 3, figsize = (8, 5))

vecX = np.linspace(-10, 10, 100)
vecY = vecX ** 2
plt.plot(vecX, vecY)
plt.show()
```

```python
plt.figure(num = 3, figsize = (8, 5))
plt.xlim((0, 5))
plt.ylim(-5, 30) # Auto tuple
plt.yticks([-10, 0, 15, 30])
plt.plot(vecX, vecY)
plt.xlabel('On Positive Axe')
plt.show()
```

```python
# Basic styling
plt.figure(num = 3, figsize = (8, 5))
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
plt.xlim((0, 5))
plt.ylim(-5, 30) # Auto tuple
plt.yticks([-10, 0, 15, 30])
plt.plot(vecX, vecY)
plt.xlabel('On Positive Axe')
plt.show()
```

```python
x = np.linspace(-1, 2, 50)
y = x - 1
plt.figure(num = 3, figsize = (8, 5))
plt.plot(x, y, linewidth = 1.0, linestyle = "dashed")
plt.xlim(-1, 2)
plt.ylim(-2, 2)
plt.xticks([-1, -0.5, 1], ['lost', 'far', 'pure'])
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
plt.show()
```
